```python
import folium
import folium.plugins # The Folium Javascript Map Library
```

```python
locations = pd.read_csv("locations.csv")
locations = locations.drop("country", axis=1)
print(len(df))
#print(locations.loc[locations.name] == "South Sudan")

def f(row):
    location = locations.loc[locations.name == row['Country']]
    return pd.Series([location['latitude'].iloc[0], location['longitude'].iloc[0]])

df[['latitude', 'longitude']] = df.apply(f, axis=1)
df
```

```python
NY_COORDINATES = (40.785091, -73.968285)
sf_map = folium.Map(location=NY_COORDINATES, zoom_start=13)
locs = df[['latitude', 'longitude']].astype('float').dropna().to_numpy()
heatmap = folium.plugins.HeatMap(locs.tolist(), radius = 10)
sf_map.add_child(heatmap)
```
